# Actividad 17. Gráfico de pay

```python
data=pd.read_csv("C:\\Users\\aportillo\Downloads\\athlete_events.csv")
data
```

```python
#Filtrar por año quitando los que no tienen datos
año=data[data["Year"].notna()]
año
```

```python
#Para ordenar desde el año más reciente
datasort=año.sort_values("Year", ascending=False)
datasort
```

```python
#Sacar solo valores del año 2016
act=datasort[datasort['Year']==2016]
act
```

```python
ax=act['ID'][:10].plot.pie(figsize=(11,6), autopct="%0.1f %%") #13688 datos en 2016
```

```python
fig=ax.get_figure()
fig.savefig("Act17-Pay.png")
```
