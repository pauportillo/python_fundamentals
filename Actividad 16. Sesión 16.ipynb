```python
data=pd.read_csv("C:\\Users\\aportillo\Downloads\\athlete_events.csv")
data
```

```python
#Ordenar que salgan solo las filas con medallas
Medallas=data[data["Medal"].notna()]
Medallas
```

```python
cantidad=Medallas[["Year", "Season", "Medal", "NOC"]]
cantidad.head(2)
```

```python
c2=pd.get_dummies(cantidad[["Medal"]])
c2.head(2)
```

```python
result=pd.concat([cantidad,c2], axis=1)
result.head(2)
```

```python
cantidad2=result.groupby(["Year", "Medal", "NOC"]).sum()
cantidad2
```

```python
cantidad2.plot(figsize=(15,5))
```

```python
cantidad2.hist(figsize=(15,10))
```

```python
cantidad2["Total_Medals"]=cantidad2["Medal_Bronze"]+cantidad2["Medal_Gold"]+cantidad2["Medal_Silver"]
cantidad2
```

```python
paises_medalleros=Medallas[['NOC','Medal']]
paises_medalleros['NOC'].value_counts().head(10).plot(kind="bar")
```
